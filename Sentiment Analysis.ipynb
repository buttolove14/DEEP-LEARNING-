```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer

```

```python


nltk.download('punkt')
nltk.download('vader_lexicon')


```

```python

# Sample social media data
social_media_data = [
    "I love this product! It's amazing.",
    "Terrible customer service. Will never buy again!",
    "Great experience with the new feature.",
    "Not happy with the quality. Disappointed."
]

```

```python
# Preprocess the data
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return ' '.join(filtered_tokens)
```

```python


# Perform sentiment analysis
def analyze_sentiment(text):
    sid = SentimentIntensityAnalyzer()
    score = sid.polarity_scores(text)
    return score

```

```python


# Analyze sentiment for each text in social media data
for text in social_media_data:
    preprocessed_text = preprocess_text(text)
    sentiment_score = analyze_sentiment(preprocessed_text)
    print(f"Text: {text}")
    print(f"Sentiment Score: {sentiment_score}")
    print()
```
