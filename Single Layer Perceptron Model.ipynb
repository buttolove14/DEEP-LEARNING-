IMPORTING DATASET FROM LIBRARY

```python
from sklearn.datasets import load_iris
```

IMPORTING PERCEPTRON  MODEL

```python
from sklearn.linear_model import Perceptron
```

IMPORTING A PERFORMANCE METRIC

```python
from sklearn.metrics import accuracy_score
```

LOADING DATASET

```python
iris = load_iris()
```

```python
print(iris)
```

INDEPENDENT AND DEPENDENT VARIABLES

```python
x = iris.data[:,(0,3)]
y = (iris.target == 0).astype(int)
```

```python
x
```

```python
y
```

APPLYING THE MODEL

```python
ptron = Perceptron(random_state = 42)
```

FITTING THE MODEL

```python
ptron.fit(x,y)
```

PREDICTING USING MODEL

```python
y_pred = ptron.predict(x)
```

```python
y_pred
```

ACCURACY OF THE MODEL

```python
print(f'Accuracy Score : {accuracy_score(y,y_pred)}')
```

COEFFICIENT VALUES FROM MODEL

```python
print(f'Coeffiecient Values : {ptron.coef_}')
```

INTERCEPT VALUE FROM THE MODEL

```python
print(f'Intercept Value : {ptron.intercept_}')
```
