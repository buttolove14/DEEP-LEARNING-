```python
def gradient_descent(gradient,start,learning_rate , iteration=50,tol=1e-06):
  vector = start
  for _ in range(iteration):
      diff = -learning_rate * gradient(vector)
      if np.all(np.abs(diff) <= tol):
         break
      vector += diff
  return vector
```
