```python
import tensorflow as tf
```

```python
con_ten = tf.constant([[1,2],[3,4]])
print('Constant Tensor :\n',con_ten)
```

```python
var_ten = tf.Variable([[1,2],[3,4]])
print('Variable Tensor :\n',var_ten)
```

```python
row_con = tf.concat(values=[con_ten , var_ten], axis = 0)
print('Row Concatenate :\n',row_con)
```

```python
col_con = tf.concat(values=[con_ten , var_ten], axis = 1)
print('Col Concatenate :\n',col_con)
```

```python
zeros = tf.zeros(shape=[3,4,4],dtype=tf.int32)
print("Zero's Tensor : \n",zeros)
```

```python
ones = tf.ones(shape=[3,4,4],dtype=tf.int32)
print("One's Tensor : \n",ones)
```
