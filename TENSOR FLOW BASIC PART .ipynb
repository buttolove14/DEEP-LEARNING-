```python
con1=tf.constant([[1.4,2.1] ,[3,4.7]])
con2=tf.constant([[5],[2]])
con3=tf.constant([[5,4],[2,1]])
```

```python
print('Tensor con1:' , con1)
print(con2)
print(con3)
```

```python
con1=tf.constant([[1.4,2.1,6] ,[3,4.7,5],[4,6,7]])
con2=tf.constant([[5,4,6],[2,7,8],[4,6,2]])
con3=tf.constant([[5,4,6],[2,1,5],[4,8,6]])
```

```python
print('Tensor con1:' , con1)
print(con2)
print(con3)
```

TRANSPOSE

```python
trans = tf.transpose(con2)
print('con2 Transpose :', trans)
```

DATATYPE CHANGE

```python
con4 = tf.cast(con1,tf.float32)
con5 = tf.cast(con2,tf.float32)
```

```python
print('con4 Tensor:',con4)
print('con5 Tensor:',con5)
```

MULTIPLICATION ELEMENTS

```python
mul_elem = tf.multiply(con4,con5)
print('Element Multiplication :' , mul_elem)
```
